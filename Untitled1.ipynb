```python
import numpy as np
import pandas as pd
from keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Preprocess the data
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32') / 255
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32') / 255
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)



# Split data for classical ML models
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)
```

```python
import matplotlib.pyplot as plt


# Function to plot images in a grid
def plot_images(images, labels, rows=2, cols=5):
    fig, axes = plt.subplots(rows, cols, figsize=(10, 5))
    axes = axes.ravel()
    
    for i in np.arange(0, rows*cols):
        axes[i].imshow(images[i], cmap='gray')
        axes[i].set_title(labels[i])
        axes[i].axis('off')
    
    plt.subplots_adjust(hspace=0.5)
    plt.show()

# Display first 10 images from the training dataset
plot_images(X_train, y_train, rows=2, cols=5)
```

```python
# Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train_flat, y_train.argmax(axis=1))

# Evaluate Random Forest
y_pred_rf = rf_model.predict(X_test_flat)
accuracy_rf = accuracy_score(y_test.argmax(axis=1), y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")
```

```python
# Train SVC with linear kernel
svc_linear = SVC(kernel='linear')
svc_linear.fit(X_train_flat, y_train.argmax(axis=1))

# Evaluate SVC with linear kernel
y_pred_svc_linear = svc_linear.predict(X_test_flat)
accuracy_svc_linear = accuracy_score(y_test.argmax(axis=1), y_pred_svc_linear)
print(f"SVC (Linear Kernel) Accuracy: {accuracy_svc_linear:.4f}")

# Train SVC with RBF kernel
svc_rbf = SVC(kernel='rbf')
svc_rbf.fit(X_train_flat, y_train.argmax(axis=1))

# Evaluate SVC with RBF kernel
y_pred_svc_rbf = svc_rbf.predict(X_test_flat)
accuracy_svc_rbf = accuracy_score(y_test.argmax(axis=1), y_pred_svc_rbf)
print(f"SVC (RBF Kernel) Accuracy: {accuracy_svc_rbf:.4f}")
```

```python
# Define MLP model
mlp_model = Sequential()
mlp_model.add(Flatten(input_shape=(28, 28, 1)))
mlp_model.add(Dense(512, activation='relu'))
mlp_model.add(Dense(256, activation='relu'))
mlp_model.add(Dense(10, activation='softmax'))

# Compile MLP model
mlp_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train MLP model
mlp_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=200, verbose=2)

# Evaluate MLP model
scores_mlp = mlp_model.evaluate(X_test, y_test, verbose=0)
accuracy_mlp = scores_mlp[1]
print(f"MLP Accuracy: {accuracy_mlp:.4f}")
```

```python
# Define CNN model
cnn_model = Sequential()
cnn_model.add(Conv2D(32, (5, 5), input_shape=(28, 28, 1), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))
cnn_model.add(Conv2D(64, (5, 5), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))
cnn_model.add(Flatten())
cnn_model.add(Dense(1000, activation='relu'))
cnn_model.add(Dense(10, activation='softmax'))

# Compile CNN model
cnn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train CNN model
cnn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=200, verbose=2)

# Evaluate CNN model
scores_cnn = cnn_model.evaluate(X_test, y_test, verbose=0)
accuracy_cnn = scores_cnn[1]
print(f"CNN Accuracy: {accuracy_cnn:.4f}")
```

```python
results = {
    "Random Forest": accuracy_rf,
    "SVC (Linear Kernel)": accuracy_svc_linear,
    "SVC (RBF Kernel)": accuracy_svc_rbf,
    "MLP": accuracy_mlp,
    "CNN": accuracy_cnn
}

for model, accuracy in results.items():
    print(f"{model}: {accuracy:.4f}")
```
